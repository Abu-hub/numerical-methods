```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import solow
import solvers
import pwt
```

```python
def ces_output(t, k, params):
    """
    Intensive form for a CES production function.

    Arguments:

        t:      (array-like) Time.
        k:      (array-like) Capital (per person/effective person).
        params: (dict) Dictionary of parameter values.
       
    Returns:

        y: (array-like) Output (per person/effective person)

    """
    # extract params
    alpha = params['alpha']
    sigma = params['sigma']
    rho   = (sigma - 1) / sigma
    
    # nest Cobb-Douglas technology as special case
    if rho == 0:
        y = k**alpha 
    else:
        y = (alpha * k**rho + (1-alpha))**(1 / rho)
    
    return y

def marginal_product_capital(t, k, params):
    """
    Marginal product of capital with CES production function.

    Arguments:

        t:      (array-like) Time.
        k:      (array-like) Capital (per person/effective person).
        params: (dict) Dictionary of parameter values.
       
    Returns:

        mpk: (array-like) Derivative of output with respect to capital, k.

    """
    # extract params
    alpha = params['alpha']
    sigma = params['sigma']
    rho   = (sigma - 1) / sigma
    
    # nest Cobb-Douglas technology as special case
    if rho == 0:
        mpk = alpha * k**(alpha - 1)
    else:
        mpk = ((alpha * k**(rho - 1) / (alpha * k**rho + (1 - alpha))) * 
               ces_output(t, k, params))    
        
    return mpk

def equation_of_motion_capital(t, k, params):
    """
    Equation of motion for capital (per person/effective person).

    Arguments:

        t:      (array-like) Time.
        k:      (array-like) Capital (per person/effective person).
        params: (dict) Dictionary of parameter values.
       
    Returns:

        k_dot: (array-like) Time-derivative of capital (per person/effective 
               person).

    """
    # extract params
    s     = params['s']
    n     = params['n']
    g     = params['g']
    delta = params['delta']
    
    y = ces_output(t, k, params)
    
    k_dot = s * y - (n + g + delta) * k
    
    return k_dot

def solow_jacobian(t, k, params):
    """
    The Jacobian of the Solow model.
    
    Arguments:

        t:      (array-like) Time.
        k:      (array-like) Capital (per person/effective person).
        params: (dict) Dictionary of parameter values.
       
    Returns:

        jac: (array-like) Value of the derivative of the equation of 
             motion for capital (per worker/effective worker) with 
             respect to k.

    """
    # extract params
    s     = params['s']
    n     = params['n']
    g     = params['g']
    delta = params['delta']

    mpk = marginal_product_capital(t, k, params) 
    
    k_dot = s * mpk - (n + g + delta)
    
    return k_dot

def analytic_k_star(params): 
    """Steady-state level of capital (per person/effective person)."""
    # extract params
    s     = params['s']
    n     = params['n']
    g     = params['g']
    alpha = params['alpha']
    delta = params['delta']
    sigma = params['sigma']
    rho   = (sigma - 1) / sigma
    
    # nest Cobb-Douglas as special case
    if rho == 0:
        k_star = (s / (n + g + delta))**(1 / (1 - alpha))
    else:
        k_star = (((1 - alpha) * s**rho) / ((n + g + delta)**rho - alpha * s**rho))**(1 / rho)
        
    return k_star

default_params = {'alpha':0.5, 'delta':0.04, 'sigma':1.05, 'g':0.02, 
                  'n':0.01, 's':0.15, 'A0':1.0, 'L0':1.0} 
```

## Motivation

Classic theoretical models of economic growth (i.e., [Solow (1956)](http://faculty.lebow.drexel.edu/LainczC/cal38/Growth/Solow_1956.pdf), [Ramsey (1928)](http://www.csus.edu/indiv/o/onure/econ%20200A%20_%20S10/Ramsey.pdf), [Cass (1965)](http://piketty.pse.ens.fr/files/Cass1965.pdf), [Koopmans (1963)](http://cowles.econ.yale.edu/P/cd/d01b/d0163.pdf)) predict that economies eventually converge to a "balanced growth path" where endogenous variables of the model are growing at constant rates. 

Key features of balanced growth:

* Constant factor shares
* Constant capital-output ratio

What do I mean by factor shares?

$$ \epsilon_K(t) \equiv \frac{r(t)K(t)}{Y(t)} \equiv \frac{\text{return to capital}\times\text{stock of capital}}{\text{GDP}}$$ 
$$ \epsilon_L(t) \equiv \frac{W(t)L(t)}{Y(t)} \equiv \frac{\text{wages}\times\text{labor supply}}{\text{GDP}}$$ 

Empirical studies of economic growth based on these models *implicity* assume that observed data are generated by economies that have converged to their respective balanced growth paths.

## Hypotheses

* Models with "balanced growth paths" not a good approximation to reality. 

* "Balanced growth" is a good approximation, but some (most) countries have simply not yet converged to their respective BGPs (i.e., they are out of steady-state).

Under either hypothesis, understanding how much of the observed growth experience can be explained by out of steady-state dynamics generated by models with balanced growth paths provides important benchmark. 

Non-existence of balanced growth paths very reasonable position to take. [Uzawa (1961)](http://www.aea-net.org/assa/2006/0106_1430_0702.pdf) places tight restrictions on form of aggregate production function required to insure balanced growth (basically growth in technology must be labor augmenting).

Uzawa's results are quite general:

* Require only that $F$ has constant-returns in $K$ and $L$.
* No assumptions on  market structure
* No assumptions on equilibrium behavior of agents
* No assumption of market-clearing

## The Solow (1956) Model

Seminal neoclassical model of economic growth via capital accumulation. Key ingredients:

* Exogenous growth in population and technology/productivity.

* Single consumption good produced by profit maximizing firms using capital and labor inputs.

* Labor and capital inputs provided by "boundedly-rational" households.

* All markets clear (i.e., supply equals demand!) at all times.

## Production

Large number of potentially *heterogenous* firms. However, so long as I am willing to assume that...

1. Factor markets for capital and labor inputs, and the market for the output good(s) are competitive: 
    * Firms are "price takers" (i.e., individual firm is small relative to size of overall market); 
    * Product homogeneity (i.e., rules out market power from product differentiation);
    * There is free entry/exit (i.e., firms can start up and shut down at will).
2. There are no *production externalities* between firms: rules out important (and realistic) features of modern economy such as like supply chains, financial interlinkages, etc.

...then the economy admits a representative firm. I am **not** arguing that firm heterogeneity is uninteresting or unimportant, but simply wish to clearly state the assumptions and their direct implications.  

## Production, cont'd

Representative firm combines inputs of capital, $K$, labor, $L$, and technology $A$, to produce a homogenous consumption good $Y$ via a constant elasticity of substitution (CES) production function with labor-augmenting technology.

$$ Y(t) = F(K(t), A(t)L(t)) = \bigg[\alpha K(t)^{\frac{\sigma - 1}{\sigma}} + (1-\alpha) (A(t)L(t))^{\frac{\sigma - 1}{\sigma}}\bigg]^{\frac{\sigma}{\sigma - 1}} \tag{1.1} $$

where $0 < \alpha < 1$ and $\sigma > 0$. The CES production technology is popular because it nests several interesing special cases. 

Assume that the level of technology grows at a constant rate $g$:

$$A(t) = A(0)e^{gt} \tag{1.2}$$

with $A(0)$ given.

## Production, cont'd

Assume the representative firm chooses demand for capital and labor inputs in order to maximise profits taking the prices (i.e., real wage, $W$ and rental rate $r$) as given.

$$\max_{K, L}\ \Pi(K, AL) = F(K, AL) - WL - rK \tag{1.3}$$

Firm optimzation problem is **static** and therefore easy to solve. First-order conditions (using CES production function from previous slide) are:

$$ r = F_K \equiv \left(\frac{\alpha K_D^{\frac{\sigma - 1}{\sigma} - 1}}{\alpha K_D^{\frac{\sigma - 1}{\sigma}} + (1-\alpha)(AL_S)^{\frac{\sigma - 1}{\sigma}}}\right)F(K_D, AL_S) \tag{1.4}$$
$$ W = F_L \equiv \left(\frac{(1-\alpha) (AL_S)^{\frac{\sigma - 1}{\sigma} - 1}}{\alpha K_D^{\frac{\sigma - 1}{\sigma}} + (1-\alpha)(AL_S)^{\frac{\sigma - 1}{\sigma}}}\right)F(K_D, AL_S) A \tag{1.5}$$

where $K_D$ and $L_D$ are the representative firm's demand for capital and labor, respectfully.

## Households

Large number of potentially *heterogenous* households. However, so long as I am willing to assume that... 

1. Asset markets are complete: individual household's can lend and borrow capital from one another (equivalent to assuming that individual households can short capital). 
2. Households have identical marginal propensities to consume out of income (equivalently, all household's have same savings rate).

...then the economy admits a representative household. Again, I am **not** arguing that household heterogeneity is uninteresting or unimportant, just want to state assumptions for the record.  

## Household's, cont'd

Size of the representative household is $N(t)$. To allow for population growth assume that 

$$ N(t) = N(0)e^{nt} \tag{2.1}$$

with $N(0)$ given.

Flow-budget constraint for the representative household is:

$$ C(t) + I(t) = Y(t) = W(t)L(t) + r(t)K(t). \tag{2.2}$$

Capital holdings of the representative household evolve according to

$$ \dot{K}(t) = I(t) - \delta K(t) \tag{2.3}$$

with $K(0)$ given.

## Household's cont'd

Suppose that representative household uses the following "rules of thumb" when choosing labor supply and consumption demand.

$$L_S(t) = \bar{l}N(t) \tag{2.4}$$

$$C_D(t) = (1 - s)Y(t) = (1 - s)\bigg[W(t)L_S(t) + r(t)K_S(t)\bigg] \tag{2.5}$$

Combining equations 2.2-2.5 yields the equation of motion for the representative household's supply of capital:

$$ \dot{K}_S(t) = s\bigg[W(t)\bar{l}N(t) + r(t)K_S(t)\bigg] - \delta K_S(t) \tag{2.6}$$

with $K_S(0)$ given.

## Equilibrium

Impose market-clearing equilibrium in markets for capital and labor (market for consumption good clears by Walras' Law):

\begin{align} 
K_D(t) =& K_S(t) = K(t) \tag{3.1} \\ 
L_D(t) =& L_S(t) = L(t) \tag{3.2}
\end{align}

at all points in time $t$.

## Equilibrium, cont'd
Substituting equations 1.4 and 1.5 (i.e., the rental rate for capital and the real wage) into equation 2.6 yields:

\begin{align} 
\dot{K}(t) =& s\bigg[\alpha K(t)^{\frac{\sigma - 1}{\sigma}} + (1-\alpha) (A(t)\bar{l}N(t))^{\frac{\sigma - 1}{\sigma}}\bigg]^{\frac{\sigma}{\sigma - 1}} - \delta K(t),\ K(0) = K_0 \\
A(t) =& A(0)e^{gt},\ A(0)=A_0 \tag{3.3}\\
N(t) =& N(0)e^{nt},\ N(0)=N_0
\end{align}

Because technology and population are both growing, Solow model as defined by equation 3.3 is non-stationary. To make the model stationary, rewrite 3.3 in terms of capital per effective person $k(t) = \frac{K(t)}{A(t)N(t)}$.

$$\dot{k}(t) = s\bigg[\alpha k(t)^{\frac{\sigma - 1}{\sigma}} + (1-\alpha)\bigg]^{\frac{\sigma}{\sigma - 1}} - (n + g + \delta)k(t),\ k(0)=k_0 \tag{3.4}$$

## A Solow model in one slide...

Models "explains" economic growth as a process of capital-accumulation. Only state-variable is capital per effective worker, $k(t)=\frac{K(t)}{A(t)L(t)}$, which evolves according to

$$\dot{k}(t) = s\bigg[\alpha k(t)^{\frac{\sigma - 1}{\sigma}} + (1-\alpha)\bigg]^{\frac{\sigma}{\sigma - 1}} - (n + g + \delta)k(t),\ k(0)=k_0$$

where the exogenous parameters are:

* $0 < s < 1$: savings rate of the representative household;
* $n$: growth rate of population, $L(t)$;
* $g$: growth rate productivity/technology, $A(t)$;
* $0 < \delta < 1$: depreciation rate of physical capital, $K(t)$;
* $0 < \alpha < 1$: controls the relative weight of factors in production;
* $0 < \sigma$: elasticity of substitution between factors of production. 

Assume that $n + g + \delta > 0$.

## First, define the model...

```python
model = solow.Model(output=ces_output,                # intensive form of output
                    mpk=marginal_product_capital,     # rental rate of capital
                    k_dot=equation_of_motion_capital, # equation of motion for capital
                    jacobian=solow_jacobian,          # dk_dot/dk (input for solvers)
                    params=default_params)            # model parameters
```

In addition to the above listed instance attributes, each `solow.Model` object has the following class attributes:

* `pwt`: A copy of the Penn World Tables data set. 
* `steady_state`: Instance of the `steady_state` class representing the long-run, steady-state equilibrium of the model. 

The `steady_state` class has additional methods for handling and storing the steady-state values of the endogenous variables of the model, as well as for numerically computing these steady-state values in the event that closed-form expressions are unavailable. 

## ... then add the steady-state.

The steady state value of capital per effective worker is a value $k^*$ such that $\dot{k}=0$. Closed-form expression exists:

$$k^* = \left(\frac{(1 - \alpha)s^{\frac{\sigma - 1}{\sigma}}}{(n + g + \delta)^{\frac{\sigma - 1}{\sigma}} - \alpha s^{\frac{\sigma - 1}{\sigma}}}\right)^{\frac{\sigma}{\sigma - 1}}$$

```python
# create a dictionary of steady state expressions...
solow_steady_state_funcs = {'k_star':analytic_k_star}

# add the dictionary of functions to the model
model.steady_state.set_functions(solow_steady_state_funcs)

# compute the steady state values
model.steady_state.set_values()

# examine the values
model.steady_state.values
```

## Calibration 

### Step 1: Pick your favorite country. 

Countries are specified by ISO3 codes: `USA`, `GBR`, `FRA`, `SGP`, `JPN`, etc.

### Step 2: Estimate $\alpha$ and $\sigma$:

For given values of $\alpha$, and $\sigma$, and observed time series data on output $Y=$ `rgdpna`, capital stock, $K=$ `rkna`, and number of people employed, $L=$ `emp`, from the [Penn World Tables (PWT)](http://www.rug.nl/research/ggdc/data/penn-world-table) use the production function to compute the implied "Solow residual":

$$ A = \left(\frac{1}{1-\alpha}\Bigg[\left(\frac{Y}{L}\right)^{\frac{\sigma - 1}{\sigma}} - \alpha\left(\frac{K}{L}\right)^{\frac{\sigma - 1}{\sigma}}\Bigg]\right)^{\frac{\sigma}{\sigma - 1}} $$

Set $k = \frac{K}{AL}$ and then define the following residual function:

\begin{align}
residual \equiv& observed\ capital\ share - predicted\ capital\ share \\
=& (1 - \verb|labsh|) - \frac{\alpha k_{it}^{\frac{\sigma - 1}{\sigma}}}{\alpha k_{it}^{\frac{\sigma - 1}{\sigma}} + (1 - \alpha)}
\end{align}

Objective is to choose $\alpha$ and $\sigma$ to minimize deviations between observed and model-predicted shares of income/output accruing to capital.

$$ \min_{\alpha, \sigma}\ \sum residual^2$$

## Calibration, cont'd

### Step 3: Calibrate remaining parameters as follows.

* Estimate $s$ using the time-average of investment share of real GDP, `i_sh`.
* Estimate $\delta$ using the time-average of `delta_k`.
* Estimate $n$, $L(0)$ using OLS: 
    - $$\ln\ \verb|emp| = \ln L(0) + n \verb|t| + \epsilon.$$
* Estimate $g$, $A(0)$ using OLS regression: 
    - $$\ln\ \verb|atfpna| = \ln A(0) + g \verb|t| + \epsilon.$$

## Calibration, cont'd

The entire calibration routine is defined by the function `calibrate_ces` that takes a `solow.Model` object as an arugment.

```python
# calibration window
start, end = None, None

# calibrate the model
results = solow.calibrate_ces(model,                   
                              iso3_code='ESP',          
                              x0=[0.5, 1.0],         
                              bounds=[start, end], 
                              method='Nelder-Mead')    
```

```python
# compute the initial condition
k0, start = model.get_initial_condition(start=start)

# solve the initial value problem
traj = model.integrate(t0=0, 
                       y0=k0, 
                       h=1.0, 
                       T=2011 - start,
                       integrator='dopri5')
```

## Summary and extensions

Economists too quick to assume that economies are on their "balanced growth paths". The out of steady-state dynamics of Solow Model seem to be *qualitatively* consistent with the observed patterns in labor shares, capital-output ratios and growth rates of per capita income for many/most countries. But...

* Predicted level of capital-output ratio seems a bit off. 
    * Possible solution: incorporate human capital into the model.
* Savings rates vary considerably over time within a country and Solow model assumes constant savings rate. 
    * Possible solution: move to Ramsey-Cass-Koopmans framework with housholds optimally choosing savings rate.
* Growth rate of technology is exogenous. 
    * Possible solution: move to Romer model of endogenous growth.

## Consumers

Representative consumer solves the following optimization problem:

$$\max_{q_i} C \equiv \Bigg[\int_0^{n}q_i^{\frac{\sigma-1}{\sigma}}di\Bigg]^{\frac{\sigma}{\sigma - 1}} \tag{3.1} $$

subject to the following budget constraint

$$ Y = wL = \int_0^{n} p_iq_i di. \tag{3.2} $$

FOCs of this problem imply that representative consumer's demand function for good $q_i$ is

$$ q_i = \left(\frac{p_i}{P}\right)^{-\sigma}C \tag{3.3} $$

with aggregate price level $P$ is

$$P = \Bigg[\int_0^{n} p_i^{1-\sigma}di\Bigg]^{\frac{1}{1-\sigma}}. \tag{3.4} $$

## Firms

Firm producing good $q_i$ is a monopolist and therefore chooses his price, $p_i$, in order to maximize profits

$$ \max_{p_i} \pi_i = p_iq_i - wL_i \tag{3.5}$$

subject to the following constraints:

\begin{align}
q_i =& \left(\frac{p_i}{P}\right)^{-\sigma}C &(\text{consumer's demand function for good $i$}) \tag{3.6}\\
q_i =& \phi(L_i - f) &(\text{production function for good $i$}.) \tag{3.7}
\end{align}

FOC determines the optimal price:

$$ p_i = \frac{\sigma}{\sigma - 1}\frac{w}{\phi} \tag{3.8}$$

We further assume that there are no barriers to entry for firms wishing to produce good $i$. This assumption simplifies life greatly...

\begin{align}
\pi_i =& 0 \\
q_i =& (\sigma - 1) \phi f \tag{3.10}\\
L_i =& \sigma f
\end{align}

...note that none of these expressions depends on $i$.

## Equilibrium

Assume market clearing...

$$ L = \int_0^{n} L_i di = n\sigma f \implies n(L) = \frac{1}{\sigma f}L \equiv (\text{number of goods produced is a function of city size}) \tag{3.11}$$

...now we can write equilibrium real GDP as 

$$ C \equiv \Bigg[\int_0^{n}q_i^{\frac{\sigma-1}{\sigma}}di\Bigg]^{\frac{\sigma}{\sigma - 1}} = (\sigma - 1)\phi f n(L)^{\frac{\sigma}{\sigma - 1}} =  (\sigma - 1)\phi f \bigg(\frac{1}{\sigma f}L\bigg)^{\frac{\sigma}{\sigma - 1}}. \tag{3.12}$$

Note that since $\sigma > 1$, exponent on $L$ is greater than 1 $\implies$ "super-linear" scaling.

## Our contribution

Suppose that elasticity of substitution between varities isn't constant, but rather varies with the number of goods produced in a city (and therefore with population!). 

Define $\rho(n) = \frac{\sigma(n) - 1}{\sigma(n)}$. Note that  

$$ \rho'(n) = \frac{\sigma(n)\sigma'(n) - (\sigma(n) - 1) \sigma'(n)}{\sigma(n)^2} = \frac{\sigma'(n)}{\sigma(n)^2}$$

and thus $\rho'(n) > (<) 0 \iff \sigma'(n) > (<) 0$. 

Recall that we require that goods are gross substitutes (i.e., $\sigma(n) > 1$). This restriction requires that $ 0 < \rho(n) < 1$. To keep things simple, suppose that 

$$\rho(n) = \beta_0 + \beta_1 n \implies \sigma(n) = \frac{1}{1 - \rho(n)} = \frac{1}{1 - \beta_0 - \beta_1 n} \tag{3.13}$$ 


## Our contribution, cont'd

From labor market clearing condition...

\begin{align}
L =& n \sigma(n) f \\
=& n\left(\frac{1}{1 - \rho(n)}\right)f \\
=& n\left(\frac{1}{1 - \beta_0 - \beta_1 n}\right)f \tag{3.14}
\end{align}

...solving for $n$ as a function of $L$:

$$ n(L) = \frac{L(1 - \beta_0)}{f + L\beta_1} \tag{3.15}$$

After quite a bit of algebra we can derive the following expression for equilibrium per capita real GDP:

$$ \frac{C}{L} = \phi\left(\frac{f\beta_0 + L\beta_1}{f + L\beta_1}\right)\left(\frac{L(1 - \beta_0)}{f + L\beta_1}\right)^{\frac{f(1-\beta_0)}{f\beta_0 + L\beta_1}} \tag{3.16}$$

Although perhaps not obvious at a glance...

$$\lim_{L\rightarrow\infty} \frac{C}{L} = \phi $$

...thus in the limit of large cities we recover constant returns to scale (i.e., same behavior as logisitic scaling).

After quite a bit of algebra we can derive the following expression for equilibrium per capita real GDP:

$$ \frac{C}{L} = \frac{\left(\frac{\rho(n(L))}{1 - \rho(n(L))}\right) \phi f n(L)^{\frac{1}{\rho(n(L))}}}{n(L)\frac{1}{1-\rho(n(L))}f} = \rho(n(L)) \phi n(L)^{\frac{1 - \rho(n(L))}{\rho(n(L))}} = \phi\left(\frac{f\beta_0 + L\beta_1}{f + L\beta_1}\right)\left(\frac{L(1 - \beta_0)}{f + L\beta_1}\right)^{\frac{f(1-\beta_0)}{f\beta_0 + L\beta_1}}$$

$$\rho(n(L)) = \beta_0 + \beta_1  \frac{L(1 - \beta_0)}{f + L\beta_1} = \frac{\beta_0(f + L\beta_1) + \beta_1L(1 - \beta_0)}{f + L\beta_1} = \frac{f\beta_0 + L\beta_0\beta_1 + \beta_1L - \beta_0\beta_1L}{f + L\beta_1} = \frac{f\beta_0 + L\beta_1}{f + L\beta_1}$$

Note that $0 < \rho(L) < 1 \implies 0 < \beta_0 < 1$.

$$1 - \rho(L) = 1 - \frac{f\beta_0 + L\beta_1}{f + L\beta_1} = \frac{f + L\beta_1 - f\beta_0 - L\beta_1}{f + L\beta_1} = \frac{f(1 - \beta_0)}{f + L\beta_1}$$
