## Part b) (10 points)

```python
# solve the model via linearization around the deterministic steady state...
!dynare++ --order 1 --no-centralize --seed 42 --check PE --per 500 --check-num 100 rbc_benchmark_HW.mod

# convert the output into a Python dictionary
rbc_hw = io.loadmat('rbc_benchmark_HW.mat')
```

## Part c) (15 points)
Using code provided in the lab, plot impulse response functions (IRFs) for output, consumption, and investment in response to a negative productivity shock. Discuss the economic mechanisms at work. You may wish to refer to the discussion of IRFs in Romer Chapter 5 and/or your lecture notes.

```python
Y_bar, I_bar, W_bar, L_bar, zero_profit_bar, K_bar, Z_bar, C_bar, r_bar = rbc_hw['dyn_ss'] 
plt.figure(figsize=(8,6))

# irfs for output, investment, and consumption in response to negative productivity shock 
plt.plot(100 * rbc_hw['dyn_irfm_eps_z_mean'][0,:100]/Y_bar , label='$Y$')
plt.plot(100 * rbc_hw['dyn_irfm_eps_z_mean'][1,:100]/I_bar , label='$I$')
plt.plot(100 * rbc_hw['dyn_irfm_eps_z_mean'][7,:100]/C_bar , label='$C$')

plt.xlabel('Periods', fontsize=15, family='serif')
plt.ylabel('% deviations', fontsize=15, family='serif')
plt.title("Output, consumption, and investment IRFs", fontsize=20, family='serif')
plt.legend(loc='best', frameon=False, prop={'family':'serif'})
plt.grid()

plt.show()

plt.figure(figsize=(8,6))

plt.plot((rbc_hw['dyn_irfm_eps_z_mean'][7,:100]+C_bar)/(rbc_hw['dyn_irfm_eps_z_mean'][0,:100]+Y_bar) , label='C/Y')

plt.xlabel('Periods', fontsize=15, family='serif')
plt.ylabel('% deviations', fontsize=15, family='serif')
plt.title("Output, consumption, and investment IRFs", fontsize=20, family='serif')
plt.legend(loc='best', frameon=False, prop={'family':'serif'})
plt.grid()

plt.show()
```

## Part e) (10 points)

Using code provided in the lab, construct the correlation matrix for the model's endogenous variables. What is the correlation between the real wage and the labor supply? How does your result compare to results reported in Romer and discussed in your lecture notes?

```python
std_vars = np.sqrt(rbc_hw['dyn_vcov'].diagonal())

rbc_hw['dyn_corr'] = rbc_hw['dyn_vcov'] / (std_vars[np.newaxis,:] * std_vars[:,np.newaxis])

np.round(rbc_hw['dyn_corr'], 2)
rbc_hw['dyn_corr'][2,3]
```
