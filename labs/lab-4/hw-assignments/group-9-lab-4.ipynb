```python


plt.figure(figsize=(8,6))

# irfs for output, investment, and consumption in response to negative productivity shock 
plt.plot(100 * rbc_order1['dyn_irfm_eps_z_mean'][0,:100] / Y_bar, label='$Y$')
plt.plot(100 * rbc_order1['dyn_irfm_eps_z_mean'][1,:100] / I_bar, label='$I$')
plt.plot(100 * rbc_order1['dyn_irfm_eps_z_mean'][7,:100] / C_bar, label='$C$')

plt.xlabel('Periods', fontsize=15, family='serif')
plt.ylabel('% deviations', fontsize=15, family='serif')
plt.title("Output, consumption, and investment IRFs", fontsize=20, family='serif')
plt.legend(loc='best', frameon=False, prop={'family':'serif'})
plt.grid()

plt.show()
```

```python
# As we can see, output declines initially due to the shock (by roughly 13 percentage points below its steady state value), then returns 
# to its initial level. Consumption falls slightly (by 5 percentage points below its steady state value), before returning to its initial level. 
# Investment falls much more than consumption (initially by 60 percentage points below its steady state value), then increases, and actually 
# goes slightly above its initial level before returning to its steady state value. This is consistent with the consumption smoothing behaviour 
# of households implied by logarithmic preferences. 
# Consumers like to maintain a stable level of consumption, so when there is a negative shock to output households devote less of
# their income to savings (ie investment) to keep consumption relatively stable. Thus investment falls much more than consumption.
# Also, note that following the initial delcline in investment, the capital stock declines, and thus the marginal product of capital rises.
# This leads households to increase their saving gradually after the shock, which is the cause of the rise in investment after its initial
# decline, even to a level above its initial value.
```

## Part d) (10 points)
Using code provided in the lab, plot IRFs for labor supply and the real wage for the model in response to a positive productivity shock.  Discuss the economic mechanisms at work. You may wish to refer to the discussion of IRFs in Romer Chapter 5 and/or your lecture notes.

```python
plt.figure(figsize=(8,6))

# percentage deviations of W and L in response to positive productivity shock
plt.plot(100 * rbc_order1['dyn_irfp_eps_z_mean'][2,:100] / W_bar, label='$W$')
plt.plot(100 * rbc_order1['dyn_irfp_eps_z_mean'][3,:100] / L_bar, label='$L$')

plt.xlabel('Periods', fontsize=15, family='serif')
plt.ylabel('% deviations', fontsize=15, family='serif')
plt.title("Real wage and labour supply IRFs", fontsize=20, family='serif')
plt.legend(loc='best', frameon=False, prop={'family':'serif'})
plt.grid()

plt.show()
```

```python
# The positive technological shock raises the marginal product of labour initially, causing the wage to jump up initially,
# peak at some time close to the shock and then gradually return to its steady state value 
# (in Walrasian markets, which are assumed, the wage is always equal to the marginal product of labour).
# Labour supply initially increases in response to the increase in the wage, as the substitution effect dominates the income effect. However,
# this increase in the wage leads to a permanent increase in lifetime income, which causes households to afford more leisure, and therefore supply
# less labour, and thus labour supply decreases, until the effects of the shock on the real wage have passed, 
# and labour supply returns to its initial value.
```
